```python
from ipywidgets import interact

def fft_exp(n_samples=10,k=0):
    t=np.arange(0,1,1/n_samples)*2*np.pi
    sig=np.sin(t)
    correlaing_sig=[np.exp(-1j*2*np.pi*k*n/n_samples) for n in range(n_samples)]
    #fft=sum([t[n]*np.exp(1j*2*np.pi*k*n/n_samples) for n in range(n_samples)])
    plt.plot(t,sig,label='signal',marker='o')
    plt.plot(t,[correlaing_i.real for correlaing_i in correlaing_sig],label='correlating real',marker='o',color='red')
    plt.plot(t,[correlaing_i.imag for correlaing_i in correlaing_sig],label='correlating imag',marker='o',color='blue')
    plt.legend()
    plt.show()
    fft=sum([sig[n]*correlaing_sig[n] for n in range(n_samples)])
    plt.scatter([k],[fft.real],label='fft real',color='red')
    plt.scatter([k],[fft.imag],label='fft imag',color='blue')
    plt.legend()
    plt.xlim([0,n_samples])
    plt.ylim([-7,7])
    plt.show()

interact(fft_exp,n_samples=(1,100),k=(0,100))
```

```python
def plot_fft(y,d=None):
    n=len(y)
    if d is None:
        d=1/n
    fft_value=np.fft.fft(y)
    freq=np.fft.fftfreq(n,d)
    fft_value=np.fft.fftshift(fft_value)
    freq=np.fft.fftshift(freq)
    band_width=freq[1]-freq[0]
    bins=np.arange(freq[0]-band_width/2,freq[-1]+band_width/2,band_width)
    plt.bar(freq,fft_value.real,label='real (cos)',alpha=0.8)
    plt.bar(freq,fft_value.imag,label='imag (sin)',alpha=0.8)
    plt.grid()
    plt.xlabel('Frequency')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()

plot_fft(y)
plot_fft(y2)
plot_fft(np.random.randn(128))
```

```python
import numpy as np
import matplotlib.pyplot as plt
L=np.pi*2
N=10
k1=17
if k1<N/2:
    raise ValueError('No aliasing error occurs if k<N/2')
k2=k1%(N/2)
grid=np.arange(0,L,L/N)
plt.scatter(grid,np.zeros_like(grid),c="black")
for x in grid:
    plt.axvline(x = x, color = "black", linestyle = "--")
x=np.arange(0,L,L/N*0.01)
y1=np.sin(k1*x)
y2=np.sin(k2*x)
plt.plot(x,y1,label=f'k={k1}')
plt.plot(x,y2,label=f'k={k2}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
print(f"The field with wave number {k1} is overlapped with the field with wave number {k2}.")
np.mean(np.abs(np.sin(k1*grid))-np.abs(np.sin(k2*grid)))
```
